```python
import cv2

import os
import sys

import matplotlib.pyplot as plt
```

```python
images = []

for image in os.listdir("./test-cases"):
    images.append(cv2.cvtColor(cv2.imread("./test-cases/" + image), cv2.COLOR_BGR2GRAY))

```

```python
image_11_median = cv2.medianBlur(images[11], 31)
```

```python
# fig, axes = plt.subplots(4, 4)
# create_grid(images, axes)

plt.imshow(image_11_median, cmap='gray')
plt.show()
```

```python
import numpy as np
```

```python
laplacian = np.array([[0,1,0],
                      [1,-4,1],
                      [0,1,0]])

image_11_median_sharpened = cv2.filter2D(image_11_median, -1, laplacian)
```

```python
from scipy.interpolate import interp1d

m = interp1d([-4*255, 4*255], [0, 255])

image_11_median_sharpened_normalized = m(image_11_median_sharpened)

plt.imshow(image_11_median_sharpened_normalized, cmap='gray')
```
